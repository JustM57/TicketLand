```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("impressions.—Åsv")
df['event_datetime_m'] = pd.to_datetime(df.event_datetime_m, yearfirst = True)
df['month'] = [d.month for d in df.event_datetime_m]
df['dayofweek'] = [d.isoweekday() for d in df.event_datetime_m]
df['day'] = [d.day for d in df.event_datetime_m]
df['hour'] = [d.hour for d in df.event_datetime_m]
```

```python
df.head()
```

```python
df.groupby(['id_show'])['is_clicked'].count().sort_values(ascending = False).head(50)
```

```python
df['daymonth'] = df['day'] + 31* df['month']
plt.rcParams['figure.figsize'] = (15, 4)
ax = sns.countplot(x = "daymonth", hue = "is_clicked", data = df[df.id_show == 25281])
```

```python
df[df.id_show == 25281].groupby(['month', 'day'])['is_clicked'].mean().plot(color='green',  lw=2, ls='solid')
```
